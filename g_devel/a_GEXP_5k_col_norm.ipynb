```python
# Normalize within cohort for GEXP, get the subframe
features = df_stack.iloc[:, 1:].mad().sort_values(ascending=False)[:5000].index
# df_stack = df_stack.loc[:, features]
df_stack = pd.concat( [pd.DataFrame(df_stack.Labels), df_stack.loc[:, features]] ,
                     axis = 1)   
for c in df_stack.Labels.str.split('_').str[0].unique():
    print(c)
    primary_subset = df_stack[df_stack.Labels.str.contains(c)]
    max_col_vals = primary_subset.iloc[:, 1:].max()
    X_train = (primary_subset.iloc[:, 1:] / max_col_vals)
    X_labeled = pd.concat( [pd.DataFrame(primary_subset.Labels), X_train] ,
                     axis = 1)
    X_labeled.index.name = c
    X_labeled.to_csv('data_GEXP_5k_col_norm/'+c+'_gexp_col_norm_5k.tsv',
                   sep = '\t')
    # break
```
